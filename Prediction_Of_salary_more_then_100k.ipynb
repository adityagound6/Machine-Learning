```python
salry=pd.read_csv('salary_company.csv')
salry
```

```python
df=salry.drop(['Unnamed: 0'],axis='columns')
df
```

```python
company_le=LabelEncoder()
degree_le=LabelEncoder()
job_le=LabelEncoder()
df['company_n']=company_le.fit_transform(df['company'])
df['job_n']=job_le.fit_transform(df['job'])
df['degree_n']=degree_le.fit_transform(df['degree'])
df
```

```python
y=df['salry_more_then_100k']
x=df.drop(['salry_more_then_100k','company','job','degree'],axis='columns')
```

```python
lg=LogisticRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15)
```

```python
lg.fit(x_train,y_train)
```

```python
lg.predict([[1,1,1]])
```

```python
lg.score(x_test,y_test)
```
